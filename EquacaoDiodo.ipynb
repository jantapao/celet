```python
import numpy as np
import matplotlib.pyplot as plt
from math import e
#%matplotlib 

# constantes
k = 1.38 * 10**(-23)
q = 1.6 * 10**(-19)
```

```python
# declaração da classe
class curva():
    def __init__(self, n, Is, T, vmax, r):
        self.n = n
        self.T = T + 273
        self.Is = Is
        self.r = r
        self.vmax = vmax
        k = 1.38 * 10**(-23)
        q = 1.6 * 10**(-19)
        self.vp = np.arange(start=0, stop = self.vmax, step = .001)
        self.vt = k*self.T/q
        self.Id = Is * (e**(self.vp/(n * self.vt)))
    
    def curva_diodo(self):
        plt.plot(self.vp, self.Id, label= self.toString())
        plt.legend()
        plt.grid()
        plt.axis([0, 1, 0, 2e-3])
        plt.show()

    def reta_carga(self):
        i = self.vmax/self.r
        plt.plot([0,self.vmax], [i, 0])
        plt.plot(self.vp, self.Id, label= self.toString())
        plt.legend()
        plt.grid()
        plt.axis([0, 0.7, 0, 2e-3])
        plt.show()
    
    def toString(self):
        return f'n:{self.n}, Is: {self.Is:.1E}, T: {self.T}'
```

```python
temp = 29
r = 10000 #10kohm
vt = k*(temp+273)/q

Is = 1000e-6/(e**(0.7/(2* vt))) # corrente polarização reversa
```

```python
vmax = 10

curva1 = curva(2, Is, temp, vmax, r)
curva1.curva_diodo()
curva1.reta_carga()
```
